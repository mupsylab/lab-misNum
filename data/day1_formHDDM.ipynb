```python
import pandas as pd
import matplotlib.pyplot as plt
import hddm
import kabuki
import ipyparallel

data1 = hddm.load_csv('sv02_total.csv')
data1 = data1[data1.blockType == "formal"]
data1 = data1[data1.series == 1]
data1 = data1[data1.rt > 0]
data1.head(10)
```

```python
m = hddm.HDDM(data1,
    depends_on = {
        "v": ["misNum"]
    }, 
    is_group_model = True,
    bias = True
)
m.find_starting_values()
m.sample(2000, burn=20)
```
